```python
import matplotlib.pyplot as plt


nb_epochs = 20
batch_size = 10
keep_prob = 0.5
learning_rate = 0.0004

filename = "{}-{}-{}-{}".format(nb_epochs, batch_size, learning_rate, keep_prob)
file = "../data/{}.txt".format(filename)
    
epochs = []
loss = []

with open(file) as f:
    for e, line in enumerate(f.readlines(), 1):
        epochs.append(e)
        loss.append(float(line))
    
fig, ax = plt.subplots()
ax.plot(epochs, loss)

ax.set(xlabel='epochs', ylabel='mean loss',
       title='epochs: {}, batch_size {}, learning_rate {}, keep_prob {}'
       .format(nb_epochs, batch_size, learning_rate, keep_prob))
ax.grid()

fig.savefig("{}.png".format(filename))
plt.show()
```
